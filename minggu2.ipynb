```python
#Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB
import numpy as np
kucing = img3.copy()

blue_channel = kucing[:, :, 2]
blue_img = np.zeros_like(kucing)
blue_img[:, :, 2] = blue_channel

height = blue_channel.shape[0]
width = blue_channel.shape[1]

rect_width = 85
rect_height = 85
cv.rectangle(blue_img,pt1=(0,height-rect_height),pt2=(rect_width,height),color=(0,100,255),thickness=-1)
plt.imshow(blue_img)
```

```python
#Lengkapi tulisan nama file pada file citra dari soal no.8. gunakan font, ukuran font, dan warna font yang sesuai keinginan anda
cat1 = blue_img.copy()
font2=cv.FONT_HERSHEY_SIMPLEX

text = 'cat1.jpeg'
fontScale = 1
color = (255, 255, 0)
thickness = 2
lineType = cv.LINE_AA

height, width = cat1.shape[:2]

cv.putText(cat1, text=text, org=(10, height - rect_height + 10), fontFace=font2, fontScale=fontScale, color=color, thickness=thickness, lineType=lineType)

plt.imshow(cat1)
```
